```
def decode_cipher(map, cipher):
    replaced = ""
    for char in cipher:
        replaced += map[char]
    return replaced

def decode_graph(map, graph):
    replaced_graph = {}
    for k in graph:
        key = ""
        for char in k:
            key += map[char]
        replaced_graph[key] = graph[k]
    return dict(sorted(replaced_graph.items(), key=lambda item:item[1], reverse=True))
            
def count_chars(text):
    counts = dict.fromkeys(string.ascii_uppercase, 0)
    for char in cipher:
        counts[char] += 1
    return dict(sorted(counts.items(), key=lambda item:item[1], reverse=True))

def count_digraphs(text):
    counts = {}
    for m in re.finditer(r'[A-Z](?=[A-Z])', text):
        graph = text[m.start():m.end()+1]
        counts[graph] = counts.get(graph, 0) + 1
    #Output in frequency order, highest first
    return dict(sorted(counts.items(), key=lambda item:item[1], reverse=True))
    
def count_trigraphs(text):
    counts = {}
    for m in re.finditer(r'[A-Z](?=[A-Z])(?=[A-Z])', text):
        graph = text[m.start():m.end()+2]
        counts[graph] = counts.get(graph, 0) + 2
    #Output in frequency order, highest first
    return dict(sorted(counts.items(), key=lambda item:item[1], reverse=True))
    
def one_off(graph):
    replaced_graph = {}
    for k in graph:
        caps = sum(1 for c in k if c.isupper())
        if caps == 1:
            replaced_graph[k] = graph[k]
    return replaced_graph
```
